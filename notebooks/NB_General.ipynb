**ASSOCIATION RULE ALGORTIHMS**

```python
# toplam session sayısı : 55 bine yakın, support ve confidence thresholdları seçimi önemli.

# en az 1 tane sepette yer almış olsunlar ( min_support = 0.000018 )

# bunu değiştirebilirim / ürün kaybedicem karşılığında / ancak data matrix im küçülecek.

# Örneğin en az 100 sepette olsunlar istiyorsam eğer --- min_support = 0.0018
```

```python
%%time
results = apriori(urun, min_support = 0.000017, use_colnames=True, verbose = 1, low_memory = True, max_len = 2)
```

```python
results  # en az 1 sepette  oldukları senaryo 
```

```python
results.sort_values("support", ascending = False)
```

```python
# BURDAKI min_threshold ise en az 1 sepette beraber oldukları senaryoda geçerli

# en az 10 sepette beraber olsunlar istiyorsam = 0.00018, en az 100 sepette beraber olsunlar istiyorsam = 0.0018

asc_rules = association_rules(results, metric = "support", min_threshold = 0.000017)  # en az 1 sepette beraberler.
```

```python
def find_all_pairs(df, urun_adı, n_pairs = 3):
        
    pairs = {}
    counter = 0

    for index, row in df.iterrows():
        
        if row["antecedents"] == {urun_adı}:
            
            score = row["lift"]
            
            pairs.update( { counter : row["consequents"] , f"score_{counter}" : score } )
            
            counter += 1
            
        if counter == n_pairs:
            break
        
    return pairs
```

```python
def find_all_pairs(df, urun_adı, n_pairs = 3):
    
    pairs = {}
    counter = 0

    for index, row in df.iterrows():
        
        if row["antecedents"] == urun_adı:
            
            score = row["lift"]
            
            pairs.update( { counter : list(row["consequents"])[0],  f"score_{counter}" : score  } )
            
            counter += 1
            
        if counter == n_pairs:
            break
        
    return pairs
```
